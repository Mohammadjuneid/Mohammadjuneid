# numpy

```python
import numpy as np
```

```python
A = np.array([[4, 2], [1, 3]])
```

```python
A
```

```python
# Determinant of the matrix
det = np.linalg.det(A)
print("\nDeterminant of given 2X2 matrix:", int(det))
```

```python
# nverse of the matrix
print(np.linalg.inv(A))
```

```python
# Transpose of the matrix
print(A.T)
```

```python
# Euclidean norm of the row vectors

from scipy.linalg import norm
from numpy.linalg import norm as norm_n
scipy_norm = norm(A)  # using scipy lib
print(scipy_norm)
numpy_norm = norm_n(A)  # using numpy lib
print(numpy_norm)
```

```python
# Euclidean norm of the column vectors
x = np.linalg.norm(A, axis=0)
print(x)
```

```python
# Eigenvalues and the corresponding eigenvectors

from numpy.linalg import eig
w,v=eig(A)
print('E-value:', w)
print('E-vector', v)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#image = plt.imread("/Users/mohammadjuneidk/Downloads/Assignment - Inferential statistics and Visualization-3/Assignment-1/laptop_data - apple-logo.jpg")
image = cv2.imread("apple-colorful-stripes.jpg")
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
blue_histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
red_histogram = cv2.calcHist([image], [1], None, [256], [0, 256])
green_histogram = cv2.calcHist([image], [2], None, [256], [0, 256]) 
 
plt.subplot(3,1,1)
plt.title("histogram of Blue")
plt.hist(blue_histogram,color="blue")
 
plt.subplot(3,1,2)
plt.title("histogram of Green")
plt.hist(green_histogram,color="green")
 
plt.subplot(3,1,3)
plt.title("histogram of Red")
plt.hist(red_histogram,color="red")
 

plt.tight_layout()
plt.show()
```

```python
blank = np.zeros(image.shape[:2], dtype='uint8')
mask = cv2.circle(blank, (image.shape[1]//2,image.shape[0]//2), 100, 255, -1)

masked = cv2.bitwise_and(image,image,mask=mask)
cv2.imshow('Mask', masked)

plt.figure()
plt.title('Colour Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
colors = ('b', 'g', 'r')
for i,col in enumerate(colors):
    hist = cv2.calcHist([image], [i], mask, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.show()
```

```python
blue_histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
red_histogram = cv2.calcHist([image], [1], None, [256], [0, 256])
green_histogram = cv2.calcHist([image], [2], None, [256], [0, 256]) 
 
plt.subplot(3,1,1)
plt.title("Line Plot of Blue")
plt.plot(blue_histogram,color="darkblue")
 
plt.subplot(3,1,2)
plt.title("Line Plot of Green")
plt.plot(green_histogram,color="green")
 
plt.subplot(3,1,3)
plt.title("Line Plot of Red")
plt.plot(red_histogram,color="red")
 
plt.tight_layout()
plt.show()
```
